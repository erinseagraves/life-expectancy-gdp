# GDP and Life Expectancy
## Identify the relationship between between the GDP and life expectancy of six countries.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statistics as stat
```

```python
# read in the data
df = pd.read_csv('all_data.csv')
```

```python
print(df.head())
print()
print(df.info())
```

```python
# first see if we have similar amounts of data for each of the 6 countries
hue = df['Country']
sns.countplot(df, y='Country', hue='Country', palette='pastel', legend=False)
```

```python
#plot life expectancy for each country to see spread
sns.boxplot(df, x='Life expectancy at birth (years)', y='Country', hue='Country', palette='pastel', legend=False)
```

```python
#plot GDP for each country to see spread
sns.boxplot(df, x='GDP', y='Country', hue='Country', palette='pastel', legend=False)
```

```python
# obtain average life expectancy for each country
df_chile = df[df['Country'] == 'Chile']
df_china = df[df['Country'] == 'China']
df_germany = df[df['Country'] == 'Germany']
df_mexico = df[df['Country'] == 'Mexico']
df_usa = df[df['Country'] == 'United States of America']
df_zimbabwe = df[df['Country'] == 'Zimbabwe']

print('Average Life Expectancies')
print('Chile: ' + str(round(stat.mean(df_chile['Life expectancy at birth (years)']), 2)))
print('China: ' + str(round(stat.mean(df_china['Life expectancy at birth (years)']), 2)))
print('Germany: ' + str(round(stat.mean(df_germany['Life expectancy at birth (years)']), 2)))
print('Mexico: ' + str(round(stat.mean(df_mexico['Life expectancy at birth (years)']), 2)))
print('USA: ' + str(round(stat.mean(df_usa['Life expectancy at birth (years)']), 2)))
print('Zimbabwe: ' + str(round(stat.mean(df_zimbabwe['Life expectancy at birth (years)']), 2)))
```

```python
# look at the average life expectancies on a bar chart
avg_le = [78.94, 74.26, 79.66, 75.72, 78.06, 50.09]
countries = ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe']

plt.bar(countries, avg_le)
plt.title('Average Life Expectancy')
plt.ylabel('Years')
plt.show()
```

## Average Life Exceptancy

The country with the highest average life expectancy was Germany with 79.66 years. The lowest was Zimbabwe with 50.09 years. The other 4 countries had averages within 5 years of Germany's.

```python
#obtain average GDP per country
print('Average GDP')
print('Chile: ' + str(round(stat.mean(df_chile['GDP']), 2)))
print('China: ' + str(round(stat.mean(df_china['GDP']), 2)))
print('Germany: ' + str(round(stat.mean(df_germany['GDP']), 2)))
print('Mexico: ' + str(round(stat.mean(df_mexico['GDP']), 2)))
print('USA: ' + str(round(stat.mean(df_usa['GDP']), 2)))
print('Zimbabwe: ' + str(round(stat.mean(df_zimbabwe['GDP']), 2)))
```

```python
# look at the average GDP on a bar chart
avg_gdp = [169788845015.31/1e12, 4957713750000.0/1e12, 3094775625000.0/1e12, 976650625000.0/1e12, 14075000000000.0/1e12, 9062579595.06/1e12]

plt.bar(countries, avg_gdp)
plt.title('Average GDP')
plt.ylabel('Dollars (in trillions)')
plt.show()
```

## Average GDP

Since the GDP values are so large the graph isn't the best visual representation. However, we can see that the USA clearly has a significantly larger average GDP than the other countries. We can also see that Zimbabwe's average GDP is the lowest and not seen on the graph.

```python
# now we want to look at each country and see if there is a correlation between gdp and life expectancy
sns.scatterplot(data=df, x=df['Life expectancy at birth (years)'], y=df['GDP'], hue=df['Country'], palette='pastel')
```

```python
# since trends aren't clear for Zimbabwe we can look at the values on its own scatter plot
plt.scatter(df_zimbabwe['Life expectancy at birth (years)'], df_zimbabwe['GDP'])
plt.title('GDP vs Life Expectancy for Zimbabwe')
plt.ylabel('GDP (dollars)')
plt.xlabel('Life Expectancy (years)')
plt.show()
```

## Conclusions

From the graphs above, we can see that for each country represented the life expectancy and GDP have positive linear correlations.
